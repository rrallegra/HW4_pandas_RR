```python
# Reference to CSV and reading CSV into Pandas DataFrame

csv_path = "raw_data/schools_complete.csv"
schools_ratings_df = pd.read_csv(csv_path, low_memory=False)
schools_ratings_df.head(10)
```

```python
csv_path2 = "raw_data/students_complete.csv"
students_ratings_df = pd.read_csv(csv_path2, low_memory = False)
students_ratings_df.head(10)
```

```python
math_score_ave = students_ratings_df['math_score'].mean()
math_score_ave
```

```python
reading_score_ave = students_ratings_df['reading_score'].mean()
reading_score_ave
```

```python
print("Average Math Score: " + str(math_score_ave))
print("Average Reading Score: " + str(reading_score_ave))
```

```python
bins = [0, 65, 100]
group_names = ['Fail', 'Pass']
pd.cut(students_ratings_df["math_score"], bins, labels = group_names)
```

```python
students_ratings_df["Math Score Summary"] = pd.cut(students_ratings_df["math_score"], bins, labels = group_names)
students_ratings_df
```

```python
students_ratings_df["Math Score Summary"].value_counts()
```

```python
percentage_math = students_ratings_df["Math Score Summary"].value_counts() / total_students * 100
percentage_math = round(percentage_math, 2)
```

```python
percentage_math
```

```python
print("% Passing Math: " + str(percentage_math))
```

```python
pd.cut(students_ratings_df["reading_score"], bins, labels = group_names)
students_ratings_df["Reading Score Summary"] = pd.cut(students_ratings_df["reading_score"], bins, labels = group_names)
students_ratings_df
```

```python
students_ratings_df["Reading Score Summary"].value_counts()
```

```python
percentage_reading = students_ratings_df["Reading Score Summary"].value_counts() / total_students * 100
percentage_reading = round(percentage_reading, 2)
```

```python
print("% Passing Reading: " + str(percentage_reading))
```

```python
overall_percentage = round((percentage_reading + percentage_math) / 2, 2)
print("% Passing Overall: " + str(overall_percentage["Pass"]))
```
